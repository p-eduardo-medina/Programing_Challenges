```python
import numpy as np
import matplotlib.pyplot as plt

N = 3  #Número de lados del poliedro
listVect = []
alpha = 2*np.pi/N
for i in range(N): #Crear los puntos de los vèrtices
    listVect.append([np.sin(i*alpha),-np.cos(i*alpha)])
V = np.array(listVect)
```

```python
listVectNP = np.array(listVect)
listVectNP, 
```

```python
vecLong = np.linspace(-1,1,100)     #Vector para determinar la longitud de los lados
pP = []    
listVertice = []         
for lado in range(N):               #Crear las coordenadas del polinomio            
    for i,dx in enumerate(vecLong):
        pP.append( [ V[lado,0] + np.tan(alpha/2)* np.cos(alpha*lado)*dx   ,  V[lado,1] + np.tan(alpha/2)* np.sin(alpha*lado)*dx ] )
        if i == len(vecLong)-1:
            listVertice.append([V[lado,0] + np.tan(alpha/2)* np.cos(alpha*lado)*dx   ,  V[lado,1] + np.tan(alpha/2)* np.sin(alpha*lado)*dx ])
pP = np.array(pP)
listVertice = np.array(listVertice)

plt.plot(pP[:,0],pP[:,1])
plt.plot(listVertice[:,0],listVertice[:,1],'o')
plt.grid()
plt.show()
```

```python
x_0,x_f = listVertice[len(listVertice)-1][0],listVertice[0][0]
lines = pP[int(len(pP)/3):]
```

```python
import random

while True:
    x_test,y_test = random.gauss(0,2),random.gauss(1,1)
    minValue = 0
    if (x_test<lines[:,0].max() and lines[:,0].min()<x_test):
        diff = 10
        for i in range(len(lines)):
            rest = x_test-lines[i][0]
            rest = abs(rest)
            if(rest<diff):
                diff = rest
                minValue = i          
        if(y_test<lines[minValue][1]):
            break
x_test,y_test
```
